```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
a=np.random.randn(50) # 평균이 0이고 표준편차가 1인 값들을 랜덤으로 뽑아줌
```

```python
print(a.mean())
print(a.std())
```

```python
plt.hist(a,bins=25, color='blue',alpha=0.7,label='hi') # histogram( bins = 막대수, color= , alpha=투명도, label=)
plt.legend(loc= 1) # loc = [1,2,3,4]분면 or 'best', 'upper left' 
```

```python
fig, axes = plt.subplots(2,2, sharex=True,sharey=True,figsize=(12,8))   # 2,2 fig / 축 공유 share_x,share_y

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(1000),bins=20)
        axes[i,j].set_xlabel('X{}{}'.format(i,j))
        axes[i,j].set_ylabel('values')
plt.subplots_adjust(wspace=0.5,hspace=0.1)    # width, height 조절
```

```python
## index_col로 column 을 index로 불러오기
df=pd.read_csv('c:/data/kallisma1/spx.csv',index_col=0,parse_dates=True)
```

```python
df.head()
```

```python
spx = df['SPX']
```

```python
fig, ax = plt.subplots()

spx.plot(ax=ax,style='k-')  #black -

ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([100,2000])

ax.set_title("haha")
```

```python
# 그래프상에 표기
crisis_data=[
    (datetime(2007,10,11),'Peak of bull market'),
    (datetime(2008,3,12),'bear Stearns Fails'),
    (datetime(2008,9,15),'Lehman Bankruptcy')
]
```

```python
fig, ax = plt.subplots()
spx.plot(ax=ax,style='k--')

for date, event in crisis_data:
    ax.annotate(event,xy=(date,spx.asof(date)+75),
               xytext=(date, spx.asof(date)+300),
               arrowprops=dict(facecolor='red',headwidth=4,width=2,
                             headlength=4),
               horizontalalignment='left',verticalalignment='top')

ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([100,2000])

ax.set_title("haha")
```
