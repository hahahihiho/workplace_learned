```python
import pandas as pd

df = pd.read_csv('C:/data/kallisma1/gdp2018.csv')
df.head()
```

## 1.

```python
df['Rank']=df['GDP'].rank(ascending=False).astype(int)
## df.sort_values() 를 이용해도 됨
df
```

## 2.

```python
df[df['Rank']<=5][['Code','Rank']].sort_values('Rank')
```

## 3.

```python
# 안됨 or 연산자에 대해 더 알아봐야함 df[df['Code']==("KOR"or"JPN"or"CHN")]
df.loc[df['Code'].isin(["KOR","JPN","CHN"])]
df[df['Code'].isin(["KOR","JPN","CHN"])]
```

## 4. 

```python
df[df['Rank']%2==1][['Code','GDP','Rank']]
```

## 5.

```python
df[(df['Rank']%2==0) & (df['Rank']<10)].sort_values(by='Rank').drop('Nation',axis=1)
```

```python
df.Rank.map(lambda x: x*2)
df.drop(0,axis=0)
df.drop('Code',axis=1)
```
