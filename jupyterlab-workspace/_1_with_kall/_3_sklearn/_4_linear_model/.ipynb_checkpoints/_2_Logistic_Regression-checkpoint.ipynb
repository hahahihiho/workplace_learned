## 2. Linear Regression of Classfication

2.1 Logistic Regression
- C(cost)를 조절해 과대적합,과소적합 문제를 해결
- C 값이 크면 -> 과대적합

```python
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_breast_cancer()
X = iris.data
y = iris.target

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)

lgrg = LogisticRegression()
lgrg.fit(X_train,y_train)
```

```python
# Cost에 따른 coef 결과
lgrg01 = LogisticRegression(C=0.1)
lgrg1 = LogisticRegression(C=1)
lgrg10 = LogisticRegression(C=10)
lgrg01.fit(X_train,y_train)
lgrg1.fit(X_train,y_train)
lgrg10.fit(X_train,y_train)

plt.plot(lgrg01.coef_[0],label = 'C=0.1')
plt.plot(lgrg1.coef_[0],label = 'C=1')
plt.plot(lgrg10.coef_[0],label = 'C=10')
plt.legend()
plt.show()
```

```python
# 확률값을 계산해줌
dead = lgrg.predict_proba(X_test)[:,0]
alive = lgrg.predict_proba(X_test)[:,1]
plt.figure(figsize=(20,7))
plt.scatter(range(len(dead)),dead,marker='o')
# plt.plot(alive)
```

```python
dead01 = lgrg01.predict_proba(X_test)[:,0]
dead1 = lgrg1.predict_proba(X_test)[:,0]
dead10 = lgrg10.predict_proba(X_test)[:,0]

plt.figure(figsize=(20,7))

plt.scatter(range(len(dead01)),dead01,marker='o',label='C=0.1')
plt.scatter(range(len(dead1)),dead1,marker='o',label='C=1')
plt.scatter(range(len(dead10)),dead10,marker='o',label='C=10')
plt.legend()
plt.show()
```
