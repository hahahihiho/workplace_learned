### Linear Support Vector
- C값 상승 -> 약한규제

```python
from sklearn.svm import LinearSVC
from sklearn import datasets
import matplotlib.pyplot as plt
from mlxtend.plotting import category_scatter
import pandas as pd
import numpy as np
```

```python
X,y = datasets.make_blobs(random_state = 42, centers = 3)
```

```python
a=np.array([[1,2,3],[1,2,3],[1,2,3]])
b=np.array([4,5,6])
a
# np.concatenate((a,b),axis=1)
```

```python
# pd.DataFrame(X)

np.concatenate((X,y),axis=1)
```

```python
mlxtend.plotting.category_scatter(X,y,label_col='a',data=pd.DataFrame(X))
```

```python
lsvm=LinearSVC()
lsvm.fit(X,y)
lsvm.coef_
```

Support Vector Machine  
y 값에 따라 값들을 분류한뒤  
각각의 class에 따라서
$${a0+a_1x_1+a_2x_2+...=k}$$
에서 k 값이 최대가 될 수 있게

```python
line = np.linspace(-15,15,2)

fig, ax = plt.subplots()
scatter = ax.scatter(X[:,0],X[:,1],c=y,s=50)

for coef, intercept, color in zip(lsvm.coef_,lsvm.intercept_,['black','blue','red']):
    ax.plot(line,-(line*coef[0]+intercept)/coef[1],c=color,marker='o')
    
ax.set_ylim(-10,15)
ax.set_xlim(-10,8)
ax.legend(*scatter.legend_elements(), loc='best',title='class')
```

```python
line = np.linspace(-15,15,2)

fig, ax = plt.subplots()
scatter = ax.scatter(X[:,0],X[:,1],c=y,s=50)

for i,coef, intercept, color in zip(range(3),lsvm.coef_,lsvm.intercept_,['black','blue','red']):
    ax.plot(line,-(line*coef[0]+intercept)/coef[1],c=color,marker='o',label=i)
    
ax.set_ylim(-10,15)
ax.set_xlim(-10,8)
ax.legend(loc='best',title='class')
```

### 비선형 데이터 다뤄보기 SVC

```python
from sklearn.svm import SVC
```

```python
X,y = datasets.make_blobs(centers=4,random_state=229)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
y=y%2
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
# function 을 씌워서 공간을 왜곡시켜 분려해내기 위함
# 3 번째에 제곱된 데이터 삽입
X_new = np.hstack([X,X[:,1:]**2])
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
figure = plt.figure()
# 회전각도: y축 기준, z축 기준
ax = Axes3D(figure,elev=0,azim=90)
mask = y==0
ax.scatter(X_new[mask,0], X_new[mask,1], X_new[mask,2], c='b',s=60,edgecolor='k')
ax.scatter(X_new[~mask,0], X_new[~mask,1], X_new[~mask,2], c='r',marker = '^',s=60,edgecolor='k')

ax.set_xlabel('feature0')
ax.set_ylabel('feature1')
ax.set_zlabel('feature1**2')
```
