## Linear Regression
### - Ridge Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
diab_data = datasets.load_diabetes()
# print(diab_data['DESCR'])

X=diab_data['data']
y=diab_data['target']
```

```python
# 4번째 특징만 가져와서 선형화

# np.newaxis 를 주게되면 세로로(column형태로 볼 수 있)
diab_X = X[:,np.newaxis,3]
diab_X.shape

X_train = diab_X[:-20]
X_test = diab_X[-20:]
y_train = y[:-20]
y_test = y[-20:]

LR = LinearRegression()
LR.fit(X_train,y_train)
print('score',LR.score(X_train,y_train),LR.score(X_test,y_test))

#graph
plt.scatter(X_train,y_train,label='train')
plt.scatter(X_test,y_test,label='test')
plt.plot(X_train,LR.coef_*X_train+LR.intercept_,color='r',label='LR')
plt.legend()
```

## 여러개 가져와서 선형화 

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=32)
LR = LinearRegression()
LR.fit(X_train,y_train)
LR.score(X_train,y_train),LR.score(X_test,y_test)
```

### Ridge 회귀
- LR은 모델의 복잡도를 제어할 방법이 없다(no parameter)
- 추가적인 제약을 줌
- 가중치의 절대값을 작게 만듬->feature가중치를 0에 가깝게(L2규제)
- alpha 를 높이면 가중치를 0에 가깝게
- Regularization

```python
from sklearn.linear_model import Ridge
```

```python
#X,y는 위에것 사용
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=321)
rdg = Ridge(alpha=0.1)
rdg.fit(X_train,y_train)

rdg.score(X_train,y_train),rdg.score(X_test,y_test)
```

```python
# 알파 값에 따른 그래프
rdg01=Ridge(alpha=0.1)
rdg01.fit(X_train,y_train)
rdg1=Ridge(alpha=1)
rdg1.fit(X_train,y_train)
rdg10=Ridge(alpha=10)
rdg10.fit(X_train,y_train)
#graph
plt.plot(rdg01.coef_,"^",label='alpha=0.1')
plt.plot(rdg1.coef_,"^",label='alpha=1')
plt.plot(rdg10.coef_,"^",label='alpha=10')
plt.plot(LR.coef_,'o',label='LR')
plt.xlabel('coef list')
plt.hlines(0,0,len(LR.coef_))
plt.legend()
```

### Lasso
- Ridge의 대안
- Ridge 와 비슷하나 어떤계수를 0으로 산출(L1 규제)

```python
from sklearn.linear_model import Lasso
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
lss = Lasso()
lss.fit(X_train,y_train)
```

```python
lss.score(X_train,y_train),lss.score(X_test,y_test)
```

```python
# lasso 와 ridge 의 coefficient 비교
lss.coef_,rdg1.coef_
```
