## Breast Cancer
## Random Forest Analyse

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
bc_data = datasets.load_breast_cancer()
X=bc_data['data']
y=bc_data['target']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=22)
```

```python
from sklearn.ensemble import RandomForestClassifier

train_acc=[]
test_acc=[]
for i in range(1,200,5):
    RF = RandomForestClassifier(n_estimators=i,random_state=i)
    RF.fit(X_train,y_train)
    
    train_acc.append(RF.score(X_train,y_train))
    test_acc.append(RF.score(X_test,y_test))
```

```python
plt.plot(range(1,200,5),train_acc,label = 'train_acc',marker='o')
plt.plot(range(1,200,5),test_acc,label = 'test_acc',marker='o')
plt.legend()
```

```python
plt.figure(figsize=(10,10))
plt.barh(y=bc_data['feature_names'], width=RF.feature_importances_)
```
