## Breast Cancer 
## analyze by deceision tree

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
bc_data = load_breast_cancer()
```

```python
X=bc_data['data']
y=bc_data['target']
```

```python
# DT modeling
# check accuracy depend on the max
train_acc=[]
test_acc=[]
for rs in range(1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = rs)
    for i in range(1,10):
        dct_model = DecisionTreeClassifier(max_depth=i)
        dct_model.fit(X_train,y_train)

        train_acc.append(dct_model.score(X_train,y_train))
        test_acc.append(dct_model.score(X_test,y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_acc,label = 'train_accuracy')
plt.plot(test_acc,label = 'test_accuracy')
plt.legend()
plt.xlabel('max_depth')
plt.ylabel('accuracy')
```
