# 모델 평가와 성능 향상

데이터를 훈련 세트와 테스트 세트로 나누는 이유는 새로운 데이터에 모델이 얼마나 잘 일반화되는지
측정하기 위함이다. 모델이 훈련 세트에 잘 맞는 것보다, 학습 과정에 없던 데이터에 대해 얼마나 잘
예측하는가가 중요하다

## 1. Cross-validation(교차 검증)

- 일반화 성능을 평가하는 데에 트레인/테스트 데이터로 한 번 나누는 것보다 더 안정적이고
  뛰어난 통계적 평가 방법
  
- 교차 검증에서는 데이터를 여러번 반복해서 나누고 여러 모델을 학습함

- 대표적으로 k-fold cross-validation (k-겹 교차 검증)

- 예를 들어 k=5를 선택한다면,

  1) 데이터를 비슷한 크기의 부분집합 5개로 나눔
  
  2) 첫번째 모델은 1번 폴드를 테스트 데이터로 사용, 2~5 폴드를 트레인 데이터로 사용
     두번째 모델은 2번 폴드를 테스트 데이터로 사용, 1,3,4,5 폴드를 트레인 데이터로 사용
     
  3) 이와 같이 5개의 정확도 값을 얻게 됨

## 1.1 scikit-learn의 교차검증(KFold)


- model_selection 모듈의 cross_val_score 함수로 구현

```python
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# 1. load iris

iris = datasets.load_iris()
```

```python
# 2. create model object

logreg = LogisticRegression()
```

```python
# 3. cross-validation

scores = cross_val_score(logreg, iris.data, iris.target)

print("cross validation scores : {}".format(scores))
```

```python
# 3.1 cross-validation : k=5

scores_k5 = cross_val_score(logreg, iris.data, iris.target, cv=5)

print("cross validation scores : {}".format(scores_k5))
```

```python
# 3.2 cross-validation : k=5 --> 평균값

scores_k5 = cross_val_score(logreg, iris.data, iris.target, cv=5)

print("cross validation scores : {:.2f}".format(scores_k5.mean()))
```

## 2. Grid Search

- 매개변수를 튜닝하여 일반화 성능을 개선하고자 함
  
- 관심 있는 매개변수들을 대상으로 간으한 조합을 시도하여 최적의 값을 찾음

- SVM에 적용하여 gamma, C의 조합을 찾아보자

###### 커널 기법

- 비선형 특성을 추가하여 선형 모델을 사용할 경우, 특성이 많을 때 연산비용이 증가할 뿐 아니라
  데이터가 부정확해질 위험이 있다.
  
- 수학적 기법을 이용하여 새로운 특성을 임의로 만들지 않고 고차원에서 학습할 수 있다. --> 커널 기법

< 주요 parameter>

- gamma : 가우시안 커널의 반경을 결정. 

  감마 값이 크면 커널의 반경이 좁아짐
  
      --> 훈련에 각각의 데이터 포인트를 더욱 반영함 --> 복잡한 증가
  
  감마 값이 작으면 커널의 반경이 넓어짐
  
      --> 훈련에 더 넓은 면적에 걸친 데이터 포인트를 반영 --> 복잡도 감소
      
- C : cost function과 동일

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# 1. load and split iris data

iris = datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 0)

print("Train set size: {} \n Test set size: {}".format(X_train.shape[0], X_test.shape[0]))
```

```python
# 2. Gred Search

best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # train SVC using combination of gamma and C
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        
        # evaluate model
        test_score = svm.score(X_test,y_test)
        
        # restore the highest score with its parameter
        if test_score > best_score:
            best_score = test_score
            best_parameters = {'gamma' : gamma, 'C' : C}
            train_score = svm.score(X_train, y_train)
            
print('Best Score: {}'.format(np.round(best_score,3)))
print('Train Score: {}'.format(train_score))
print('parameters: {}'.format(best_parameters))
```

###### < Note>

- 위의 결과에 따라 <u>모델의 정확도가 97%라고 결론 짓는 것은 위험</u>한 것일 수 있음

- 여러가지 매개변수를 조합하여 테스트 세트의 정확도가 가장 높은 조합을 선택했으나,

- 이 정확도는 새로운 데이터에 이어지지 않을 수 있다

- 매개변수를 조정하기 위해 테스트 세트를 이미 사용했기 때문에 최종 모델의 정확성은  
  새로운 데이터가 생성될 때까지 알 수 없다.
  
- 즉, <u>최종 평가를 위해서는 모델을 만들고 튜닝할 때 사용하지 않은</u> **독립된 데이터 셋이 필요**하다.

- 애당초 데이터셋을 3개로 나누어 이 문제를 해결할 수 있다.

   | 훈 련 세 트(train data) | 검 증 세 트(validation data) | 테스트 세트(test data) |
    
    
- 그러나 이 역시 데이터의 수가 적을 때는 부정확할 수 있다.

```python
# 1, 데이터 분할 step1 --> train + validation set / test set

X = iris.data
y = iris.target

X_trainval, X_test, y_trainval, y_test =train_test_split(
    X, y, test_size = 0.25, random_state = 0)

# 2. 데이터 분할 step2 --> train / validation

X_train, X_valid, y_train, y_valid =train_test_split(
    X_trainval, y_trainval, test_size = 0.25, random_state = 1)

print("train set size: {} \n validation set size: {} \n test set size: {}".format(X_train.shape[0],
                                                                               X_valid.shape[0],
                                                                               X_test.shape[0]))
```

```python
# 2. Grid Search

best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # train SVC using combination of gamma and C
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        
        # evaluate model
        score = svm.score(X_valid,y_valid)
        
        # restore the highest score with its parameter
        if score > best_score:
            best_score = score
            best_parameters1 = {'gamma' : gamma, 'C' : C}

# train / validation set을 합쳐 최종 모델 적합 후
# test set 사용해 평가

svm = SVC(**best_parameters1)
svm.fit(X_trainval, y_trainval)
test_score = svm.score(X_test, y_test)
           
print('Best Score in validation: {}'.format(np.round(best_score,3)))
print('parameters: {}'.format(best_parameters))
print('Test Score with best parameters: {}'.format(test_score))
```

## 3. Cross-validation + Grid Search

- 데이터를 train / validation / test set으로 나누는 방법은 성능이 좋고 널리 사용되지만,

- 위에서 본 바와 같이 데이터를 나누는 방법에 매우 민감

- 일반화 성능을 더 잘 평가하기 위해서는 훈련세트와 검증세트를 한번만 나누지 않고,

- 교차 검증을 사용해서 각 매개변수의 조합의 성능을 평가할 수 있음

```python
best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # train SVC using combination of gamma and C
        svm = SVC(gamma=gamma, C=C)
        
        # cross-validation
        scores = cross_val_score(svm, X_trainval,y_trainval, cv = 5)
        
        # mean of cross-validation accuracy
        score = np.mean(scores)
        
        # restore the highest score with its parameter
        if score > best_score:
            best_score = score
            best_parameters = {'C' : C, 'gamma' : gamma}
            
## train / validation set을 합쳐 최종 모델 적합 후
# test set 사용해 평가

svm = SVC(**best_parameters)
svm.fit(X_trainval, y_trainval)
```

### ** < <u>GridSearchCV</u> > **

- 교차 검증을 사용한 그리드 서치를 매개변수 조정 방법으로 널리 사용하므로,

- scikit-learn은 GridSearchCV를 제공하고 있음

- 검색 대상이 되는 매개변수를 딕셔너리 자료형으로 구성해야 함

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 1. parameter grid 생성

param_grid = {'C' : [0.001,0.01,0.1,1,10,100],
             'gamma' : [0.001,0.01,0.1,1,10,100]}
```

```python
# 2. GridSearchCV 이용하여 객체 생성

grid_search = GridSearchCV(SVC(), param_grid, cv = 5, return_train_score=True)
```

```python
# 3. 데이터 분할
### 데이터가 과대적합 되는 것을 방지하기 위해, 또한 최종 모델의 
### 객관적인 정확도 평가를 위해 test data를 분리하고 gridsearch 
### 및 cross-validation에 사용되지 않도록 한다

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                   random_state = 0, test_size = 0.2)
```

```python
# 4. GridSearch model fitting

grid_search.fit(X_train, y_train)
```

######  ** [GridSearchCV().fit] **

fit 메서드는 최적의 매개변수를 찾는 일뿐만 아니라,  
교차검증 성능이 가장 좋은 매개변수로 전체 훈련 데이터 세트에 대해  
새로운 모델을 자동으로 만든다.  

생성된 훈련 모델에 최적 매개변수로 일반화 성능을 평가하려면,  
테스트 세트를 매개변수로 넘겨 score 메서드를 호출하면 된다.

```python
# 5. model evaluation

print("test set score : {}".format(grid_search.score(X_test, y_test)))
```

```python
# 6. best parameter and best score

print("best parameters :{}".format(grid_search.best_params_))
print("best score :{}".format(grid_search.best_score_))
```

###### ** Note **

- 교차검증과 grid search를 복합 적용한 결과, 5. model evaluation과 같이  
  높은 정확도를 얻었다.
- 하지만 5.에서의 정확도와 6.에서의 best score를 혼동해서는 안된다.
- 6.에서의 best_score_ 속성은 훈련세트에서 수행한 교차검증의 평균 정확도가  
  저장된 것이다.
- 5.에서 테스트 세트의 정확도는 교차검증과 그리드 서치의 결과로 산정한  
  최적 매개변수를 적용하여 전체 훈련데이터 셋에 대해 훈련한 최종 모델에  
  테스트 데이터를 적용했을 때, 예측이 다 맞았음을 의미
  
  < ** <u>테스트 세트는 최종 모델을 위한 그리드 서치 및 교차 검증에 사용되지 않았다</u> ** >

## 3. 평가 지표와 측정

#### 3.1 이진 분류의 평가 지표

- 이진분류에는 양성 클래스와 음성 클래스가 있으며, 양성 클래스가 우리의 관심사
- 오분류한 샘플의 수가 원하는 정보의 전부는 아니므로, 정확도(accuracy)만으로는  
  충분치 않을 수 있다.
- ex) 암 진단
  - 음성 : 건강 / 양성 : 암
  - 건강한 사람을 양성으로 진단한 경우, 환자에게 비용 손실과 불편을 가져다준다.  
    이를 false positive (거짓 양성)이라 명명 --> 1종 오류
  - 암 환자를 음성으로 진단한 경우, 치명적인 결과 초래  
    이를 false negative (거짓 음성)이라 명명 --> 2종 오류

###### * <불균형 데이터셋> *

- 두 종류의 에러는 두 클래스 중 하나가 다른 것보다 훨씬 많을 때 더 중요
- 현실에는 불균형 데이터 셋이 많음  
  예) 광고 클릭  
    - 인터넷 광고는 대부분 클릭까지 이어지지 않는다. 사용자가 클릭을 할 때까지  
      100개의 광고를 보여주고 1번의 클릭이 일어날 수도 있다. 이러한 경우,  
      99개의 '클릭 No' 음성 클래스와 1개의 '클릭 yes'이므로 데이터 셋의 클래스  
      비율은 99:1이 된다.

```python
# <불균형 데이터셋>

# 1. load_digits data : 손글씨 숫자 0-9를 분류하는 데이터

digits = datasets.load_digits()
digits.data.shape
```

```python
# 데이터 확인

print(np.bincount(digits.target))
print("0의 개수: {}".format(np.sum(digits.target==0)))
```

```python
# 2. 양성 클래스가 적은 데이터 셋 만들기

y = digits.target == 9
np.sum(y)
```

### *[<u> **모델 평가 지표** </u>]*

## - $Accuracy(정확도) = {TP + TN \over TP+TN+FP+FN}$


## - $Precision(정밀도) = {TP \over TP+FP}$


## - $Recall(재현률,sensitivity:민감도) = {TP \over TP+FN}$


## - $f1-score = {2 * Precision * Recall \over Precision + Recall}$

### - 정밀도 : 거짓 양성(FP)를 줄이는 것이 목표일 때 성능 지표로 사용  
    ex) 신약의 임상 실험
    
### - 재현률 : 거짓 음성(FN)를 줄이는 것이 목표일 때 성능 지표로 사용  
    ex) 암 진단

```python
# 위 4가지 모델의 f1-score 알아보기

from sklearn.metrics import f1_score
```

```python
# 위 4가지 모델의 f1-score 알아보기

print("빈도 기반 더미 모델 f1_score:\n{}".format(f1_score(y_test, pred_most_frequent)))
print("무작위 더미 모델 f1_score:\n{}".format(f1_score(y_test, pred_dummy)))
print("Decision Tree f1_score:\n{}".format(f1_score(y_test, pred_tree)))
print("Logistic Regression f1_score:\n{}".format(f1_score(y_test, pred_logreg)))
```

### [ROC Curve]

- ROC 곡선은 모델의 특성을 분석하는 데 널리 사용하는 도구

- 진짜 양성 비율(TPR, 재현률)에 대한 거짓 양성 비율(FPR)을 나타냄


## - Specificity(특이도) = ${TN \over TN + FP}$


## - FPR = ${FP \over TN + FP}$ = 1 - 특이도


## - TPR = ${TP \over TP + FN}$

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, logreg.decision_function(X_test))

plt.plot(fpr, tpr, label = 'ROC Curve')
plt.xlabel("FPR(1-Specificity)")
plt.ylabel("TPR(Recall)")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, logreg.decision_function(X_test))

fpr_tree, tpr_tree, thresholds_tree = roc_curve(y_test, tree.predict_proba(X_test)[:,1])

plt.plot(fpr, tpr, label = 'ROC Curve')
plt.plot(fpr_tree, tpr_tree)
plt.xlabel("FPR(1-Specificity)")
plt.ylabel("TPR(Recall)")
```

[해석]

- x축, y축이 모두 1인 경우는 모두 True로 분류한 경우
- x축, y축이 모두 0인 경우는 모두 False로 분류한 경우
- 이상적으로 완벽히 분류한 모형의 경우 x축은 0, y축은 1이 되어 AUC = 1
- AUC(Area Under the ROC Curve) 값이 클수록 더 높은 분류 성과를  
  가지는 것으로 판단

#### 1.3 Ridge

- 일반 Linear Regression은 모델의 복잡도를 제어할 방법이 없다. (파라미터가 없기 때문)
- 릿지 회귀에서의 가중치 선택은 train data에 대한 것 뿐 아니라 추가 제약조건을 만족시키는 방향으로 진행
- 방법 : 가중치의 절대값을 가능한 작게 만듦
  --> 모든 feature의 가중치를 0에 가깝게 많들어서 기울기를 최대한 작게 만든다. --> 규제
- Regularization : 모델이 과대적합되지 않도록 강제로 제한 --> L2 규제
- 매개변수 조정
  --> alpha 값을 높이면 규제를 강하게 하는 것 --> 계수를 더 0에 가깝게 만듦

```python
from sklearn.linear_model import Ridge
```

```python
# 1. 데이터 분할

X = diab.data
y = diab.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=0)
```

```python
# 2, object for modelling

rdg = Ridge()
rdg.fit(X_train, y_train)
```

```python
# 3. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(rdg.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(rdg.score(X_test, y_test)))
```

```python
# 4, parameter tuning

rdg01 = Ridge(alpha = 0.1)
rdg01.fit(X_train, y_train)
```

```python
# 5. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(rdg01.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(rdg01.score(X_test, y_test)))
```

```python
# 6, parameter tuning

rdg001 = Ridge(alpha = 0.01)
rdg001.fit(X_train, y_train)

# 7. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(rdg001.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(rdg001.score(X_test, y_test)))
```

```python
# 8, parameter tuning

rdg10 = Ridge(alpha = 10)
rdg10.fit(X_train, y_train)

# 9. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(rdg10.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(rdg10.score(X_test, y_test)))
```

#### 1.4 Lasso

- Ridge의 대안
- Ridge와 마찬가지로 모든 계수를 0에 가깝게 만듦
- 규제 방식에 차이가 있음 --> L1 규제라 명명
- 어떤 계수값은 0으로 산출 --> 완전 제외

```python
from sklearn.linear_model import Lasso
```

```python
# 1. 데이터 분할

X = diab.data
y = diab.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=0)
```

```python
# 2, modelling

lss = Lasso()
lss.fit(X_train, y_train)
```

```python
# 3. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(lss.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(lss.score(X_test, y_test)))
print("사용한 특성의 수: {}".format(np.sum(lss.coef_ != 0)))
```

```python
# 4, parameter tuning

lss001 = Ridge(alpha = 0.01)
lss001.fit(X_train, y_train)
```

```python
# 5. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(lss001.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(lss001.score(X_test, y_test)))
print("사용한 특성의 수: {}".format(np.sum(lss.coef_ != 0)))
```

```python
# 5. 훈련 결과 확인

print("훈련 세트 점수: {:.3f}".format(lss001.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(lss001.score(X_test, y_test)))
print("사용한 특성의 수: {}".format(np.sum(lss.coef_ != 0)))
```

#### 2.1 Logistic Regression

- Parameter C를 조정해 과대적합 혹은 과소적합 문제를 해결
- C(cost function) 값이 크면(높은 코스트) --> 훈련을 더 복잡하게 --> 약한 규제
- C 값이 작으면(낮은 코스트) --> 훈련을 덜 복잡하게 --> 강한 규제

```python
from sklearn.linear_model import LogisticRegression
```

###### 2.1.2 Breast Cancer

```python
# 1. 데이터 불러오기

cancer = datasets.load_breast_cancer()

# 2. 데이터 분할 - 이항분류

X = cancer.data
y = cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

# 3. modelling

logreg = LogisticRegression().fit(X_train, y_train)
logreg

# 4. 결과 비교

print("훈련 세트 점수: {:.3f}".format(logreg.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg.score(X_test, y_test)))

# 5. parameter tuning

logreg001 = LogisticRegression(C = 0.01).fit(X_train, y_train)
logreg001

logreg1000 = LogisticRegression(C = 100).fit(X_train, y_train)
logreg1000
```

```python
# 5.1 C = 0.1 결과 비교 --> 낮은 코스트 --> 훈련 덜 복잡 --> 강한 규제 --> 과소적합

print("훈련 세트 점수: {:.3f}".format(logreg001.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg001.score(X_test, y_test)))
```

```python
# 5.2 C = 10 결과 비교 --> 높은 코스트 --> 훈련 더 복잡 --> 약한 규제 --> 과대적합

print("훈련 세트 점수: {:.3f}".format(logreg100.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg100.score(X_test, y_test)))
```

```python
# 6. 그래프

plt.figure(figsize = (10,5))
plt.plot(logreg.coef_.T, "o", label = "C = 1")
plt.plot(logreg001.coef_.T, "^", label = "C = 0.01")
plt.plot(logreg100.coef_.T, "v", label = "C = 100")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
plt.hlines(0,0,cancer.data.shape[1])
plt.xlabel("feature")
plt.ylabel("coefficient")
plt.ylim(-8,8)
plt.legend()
```

```python
# 7. C는 각각 0.001, 1, 100 일 때, L1 규제를 적용하여, 각각의 훈련/테스트 정확도를 출력하고
# 그래프를 for문을 이용하여 작성해보세요.
```

```python
plt.figure(figsize = (10,5))
for C, marker in zip([0.001,1,100],['o','^','v']):
    logreg_l1 = LogisticRegression(C = C, penalty = 'l1').fit(X_train, y_train)
    print("C={:.3f}인 L1 로지스틱 회귀의 훈련 정확도: {:.2f}".format(
        C, logreg_l1.score(X_train,y_train)))
    print("C={:.3f}인 L1 로지스틱 회귀의 테스트 정확도: {:.2f}".format(
        C, logreg_l1.score(X_test,y_test)))
    plt.plot(logreg_l1.coef_.T, marker, label = "C={:.3f}".format(C))

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.xlabel("feature")
plt.ylabel("coefficient")
plt.legend(loc = 4)
```

```python
XX, YY = np.meshgrid(xx,yy)
```

```python
# 그래프 완성

linear_svm_3d = LinearSVC().fit(X_new, y)

coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

# 선형 결정경계

figure = plt.figure()
ax = Axes3D(figure, elev=-152,azim=-25)

xx = np.linspace(X_new[:,0].min()-2, X_new[:,0].max()+2, 50)
yy = np.linspace(X_new[:,1].min()-2, X_new[:,1].max()+2, 50)

XX, YY = np.meshgrid(xx,yy)
ZZ = (coef[0]*XX + coef[1]*YY+intercept)/(-coef[2])

ax.plot_surface(XX, YY, ZZ, rstride = 8, cstride = 8, alpha = 0.3, color='black',
                    cmap = cm.Accent)
ax.scatter(X_new[mask,0], X_new[mask,1],X_new[mask,2], c='b',s=60, edgecolor='k')
ax.scatter(X_new[~mask,0], X_new[~mask,1],X_new[~mask,2], c='r',
           marker = '^', s=60, edgecolor='k')
ax.set_xlabel('feature0')
ax.set_ylabel('feature1')
ax.set_zlabel('feature0 ** 2')
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX,YY,dec.reshape(XX.shape), levels = [dec.min(), 0, dec.max()],
            alpha = 0.5)
plt.scatter(X[:,0], X[:,1], c = y, s = 50, edgecolor = 'b')
```

###### 커널 기법

- 비선형 특성을 추가하여 선형 모델을 사용할 경우, 특성이 많을 때 연산비용이 증가할 뿐 아니라
  데이터가 부정확해질 위험이 있다.
  
- 수학적 기법을 이용하여 새로운 특성을 임의로 만들지 않고 고차원에서 학습할 수 있다. --> 커널 기법

** 주요 parameter

- gamma : 가우시안 커널의 반경을 결정. 

  감마 값이 크면 커널의 반경이 좁아짐
  
      --> 훈련에 각각의 데이터 포인트를 더욱 반영함 --> 복잡한 증가
  
  감마 값이 작으면 커널의 반경이 넓어짐
  
      --> 훈련에 더 넓은 면적에 걸친 데이터 포인트를 반영 --> 복잡도 감소
      
- C : cost function과 동일

###### 2.2.1 Breast Cancer

```python
from sklearn.svm import SVC
```

```python
# 1. load and split data

cancer = datasets.load_breast_cancer()

X = cancer.data
y = cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# 2. modelling

svc = SVC()
svc.fit(X_train, y_train)
```

```python
# 3, 결과 비교

print("훈련세트 정확도: {:.3f}".format(svc.score(X_train, y_train)))
print("테스트세트 정확도: {:.3f}".format(svc.score(X_test, y_test)))
```

```python
# 4.1 parameter tuning --> gamma = 0.1

svc_gamma0001 = SVC(gamma=0.0001).fit(X_train, y_train)
```

```python
# 4.1.1 결과

print("훈련세트 정확도: {:.3f}".format(svc_gamma0001.score(X_train, y_train)))
print("테스트세트 정확도: {:.3f}".format(svc_gamma0001.score(X_test, y_test)))
```

```python
# 5. 정규화
```

```python
# 1) 변수들의 값의 대략적인 크기 확인

plt.boxplot(X_train, manage_xticks=False)
plt.yscale("symlog")
plt.xlabel("features")
plt.ylabel("value")
```

```python
# 2) 모든 특성 정규화

# 2-1) 트레인 데이터 특성별 최소값/최대값 추출
X_train_min = X_train.min(axis=0)
X_train_max = X_train.max(axis=0)

# 2-2) 트레인 데이터 특성별 범위

X_train_range = X_train_max - X_train_min

# 2-3) 트레인 데이터 정규화

X_train_scaled = (X_train-X_train_min)/X_train_range

# 2-4) 정규화 확인

print("특성별 최소값\n {}".format(X_train_scaled.min(axis=0)))
print("특성별 최소값\n {}".format(X_train_scaled.max(axis=0)))

# 2-5) 테스트 데이터 정규화

X_test_scaled = (X_test-X_train_min)/X_train_range
```

```python
# 3) modelling

svc_scaled = SVC().fit(X_train_scaled, y_train)
```

```python
# 4) 결과

print("train accuracy\n {:.3f}".format(svc_scaled.score(X_train_scaled, y_train)))
print("test accuracy\n {:.3f}".format(svc_scaled.score(X_test_scaled, y_test)))
```

```python
# 5) tuning

svc_scaled_gamma1 = SVC(gamma=1).fit(X_train_scaled, y_train)
```

```python
# 5) tuning result

print("train accuracy\n {:.3f}".format(svc_scaled_gamma1.score(X_train_scaled, y_train)))
print("test accuracy\n {:.3f}".format(svc_scaled_gamma1.score(X_test_scaled, y_test)))
```

```python
# 1. Create toy data

X, y = datasets.make_blobs(n_samples = 100, random_state = 4)

# 2. split data

X_train, X_test = train_test_split(X, test_size = 0.2)
```

```python
# 3. Scaling

# 3-1) apply range of train set
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3-2) use each scaler
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)

X_test_scaled_each = test_scaler.transform(X_test)
```

```python
# 4. 그래프 작성

fig, axes = plt.subplots(1,3,figsize = (15,4))

# 1번 그래프 : 원본 데이터 (스케일 조정전)

axes[0].scatter(X_train[:,0], X_train[:,1], c='red', 
               label = "train data", s=50)
axes[0].scatter(X_test[:,0], X_test[:,1], c='blue', 
               label = "test data", s=50)

axes[0].legend(loc = 'upper left')
axes[0].set_title("before scaling", size=20)


# 2번 그래프 : 원본 데이터 (스케일 조정전)

axes[1].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c='red', 
               label = "train data", s=50)
axes[1].scatter(X_test_scaled[:,0], X_test_scaled[:,1], c='blue', 
               label = "test data", s=50)
axes[1].set_title("right scaling", size=20)

# 3번 그래프 : 원본 데이터 (스케일 조정전)

axes[2].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c='red', 
               label = "train data", s=50)
axes[2].scatter(X_test_scaled_each[:,0], X_test_scaled_each[:,1], c='blue', 
               label = "test data", s=50)
axes[2].set_title("wrong scaling", size=20)

for ax in axes:
    ax.set_xlabel("feature0")
    ax.set_ylabel("feature1")
    
```
